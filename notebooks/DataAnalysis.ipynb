```python
import numpy as np
import logging
from mlp.data_providers import MNISTDataProvider
import matplotlib.pyplot as plt
import pickle
%matplotlib notebook
```

```python
file_name = 'nonlins1'
fileObject2 = open(file_name,'rb');
length = 5
array = []
for i in range(0,length):
    array.append(pickle.load(fileObject2));
fileObject2.close()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8,8))
subarray = array[0:2]
for net in subarray:
    axs[0].plot(np.arange(1, net[1][0].shape[0]) * 5, net[1][0][1:, net[1][1]['error(valid)']], label=net[0])
    axs[0].plot(np.arange(1, net[1][0].shape[0]) * 5, net[1][0][1:, net[1][1]['error(train)']], label=net[0])
    axs[0].set_ylabel('Error')
    axs[0].legend()
    axs[1].plot(np.arange(1, net[1][0].shape[0]) * 5, net[1][0][1:, net[1][1]['acc(valid)']], label=net[0])
    axs[1].plot(np.arange(1, net[1][0].shape[0]) * 5, net[1][0][1:, net[1][1]['acc(train)']], label=net[0])
    axs[1].set_xlabel('Epochs')
    axs[1].set_ylabel('Accuracy')
    axs[1].legend()
```
