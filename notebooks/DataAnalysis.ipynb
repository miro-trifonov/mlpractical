```python
import numpy as np
import logging
from mlp.data_providers import MNISTDataProvider
import matplotlib.pyplot as plt
import pickle
%matplotlib notebook
```

```python
file_name = 'exp_conv_1'
file_name2 = 'exp_batch_50'
fileObject2 = open(file_name,'rb');
fileObject3 = open(file_name2,'rb');
length = 1
array = []
for i in range(0,length):
    array.append(pickle.load(fileObject2));
for i in range(0,length):
    array.append(pickle.load(fileObject3));
fileObject2.close()
```
