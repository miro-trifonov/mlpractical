```python
import numpy as np
import logging
from mlp.data_providers import MNISTDataProvider
import matplotlib.pyplot as plt
import pickle
%matplotlib notebook
```

```python
file_name = 'exp_conv_1'
# file_name2 = 'exp_bathch_norm_2'
fileObject2 = open(file_name,'rb');
# fileObject3 = open(file_name2,'rb');
length = 1
array = []
for i in range(0,length):
    array.append(pickle.load(fileObject2));
# for i in range(0,length):
#     array.append(pickle.load(fileObject3));
fileObject2.close()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8,8))
# subarray = [array[1]] + [array[3]]
i = 0
# this = ["ReLU model, two hidden layers","Sigmoid model, 2 hidden layers"]
for net in array:
#     axs[0].plot(np.arange(1, net[1][0].shape[0]) * 5, net[1][0][1:, net[1][1]['error(train)']], )
    axs[0].plot(np.arange(1, net[1][0].shape[0])*2, net[1][0][1:, net[1][1]['acc(train)']],label=(net[0]))
    axs[0].set_ylabel('Acc(train)')
    axs[0].legend()
#     axs[2].plot(np.arange(1, net[1][0].shape[0]) * 5, net[1][0][1:, net[1][1]['acc(train)']])
    axs[1].plot(np.arange(1, net[1][0].shape[0]) * 2, net[1][0][1:, net[1][1]['acc(valid)']])
    axs[1].set_xlabel('Epochs')
    axs[1].set_ylabel('Accuracy(valid)')
    axs[1].legend()
    i+= 1
    plt.savefig('{0}.pdf'.format("learning_rate"))
```
